```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

# **Question 2**

**Five Visualizations for the Student Performance Dataset**

```python
#Load the dataset
student_data = pd.read_csv('StudentsPerformance.csv')
student_data.head()
```

### **Gender Distribution (Pie chart)**

```python
# Pie chart for gender distribution
gender_counts = student_data['gender'].value_counts()

# Plot pie chart
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightcoral'], startangle=90)
plt.title('Gender Distribution')
plt.show()
```

**The pie chart allows us to see the proportional representation of genders. This type of visualization makes it easier to quickly understand gender balance in the dataset by displaying percentages. For example, it helps to see if the dataset is equally distributed between male and female students, which is important for identifying potential gender biases in performance analysis.**

### **Race/Ethnicity Distribution (Bar Plot)**

```python
# Horizontal bar plot for race/ethnicity distribution
sns.countplot(y='race/ethnicity', data=student_data, palette='Set3')
plt.title('Race/Ethnicity Distribution')
plt.ylabel('Race/Ethnicity')
plt.xlabel('Number of Students')
plt.show()
```

**The horizontal bar plot is useful for comparing the distribution of students across different racial/ethnic groups. By using horizontal bars, longer labels (like race categories) are easier to read, and it provides a quick visual for assessing the diversity of the student population. This is especially useful when analyzing if different racial/ethnic groups have an impact on academic performance.**

### **Lunch Type vs Average Score**

```python
# Swarm plot for lunch type vs average score
sns.swarmplot(x='lunch', y='average_score', data=student_data, palette='husl')
plt.title('Lunch Type vs Average Scores')
plt.xlabel('Lunch Type')
plt.ylabel('Average Test Scores')
plt.show()
```

**The strip plot shows individual student scores based on parental education levels, making it easier to identify outliers and clusters. This visualization is helpful for identifying whether certain parental education levels tend to have more students with higher or lower average scores, or if there's a wide variation in performance for each category.**

### **Parental Education vs Average Scores (Strip plot)**


```python
# Create a new column 'average_score' as the mean of math, reading, and writing scores
student_data['average_score'] = student_data[['math score', 'reading score', 'writing score']].mean(axis=1)


# Strip plot for parental education vs average score
sns.stripplot(x='parental level of education', y='average_score', data=student_data, jitter=True, palette='Dark2')
plt.title('Parental Level of Education vs Average Scores')
plt.xlabel('Parental Education Level')
plt.ylabel('Average Test Scores')
plt.xticks(rotation=45)
plt.show()
```

**The swarm plot visualizes each student’s average score based on their lunch type, making it possible to observe how individual performances vary between students on free/reduced lunch and those on standard lunch. This visualization provides more detail than a bar or box plot because it shows every data point, which makes it easy to spot patterns or outliers.**

### **Test Preparation vs Performance (Violin Plot)**

```python
# Violin plot for test preparation course vs average score
sns.violinplot(x='test preparation course', y='average_score', data=student_data, palette='coolwarm')
plt.title('Test Preparation Course vs Average Scores')
plt.xlabel('Test Preparation Course')
plt.ylabel('Average Test Scores')
plt.show()
```

**The violin plot provides an in-depth look at the distribution of scores for students who took the test preparation course and those who didn’t. It shows not only the range of scores but also the density (how concentrated the scores are) for each group, which can help in understanding whether the test preparation course has a noticeable effect on student performance.**
